```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Menu Ds
df_menu = pd.read_csv('./Menu.csv')
df_menu

columns = ['id','sponsor','physical_description','notes','date', 'location', 'status', 'page_count', 'dish_count']
df_menu_filtered = df_menu[columns]

top_sponsors = df_menu_filtered.sponsor.value_counts().head(15)

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x=top_sponsors.index, y=top_sponsors);
```
