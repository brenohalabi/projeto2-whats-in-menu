```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Menu Ds
df_dish = pd.read_csv('./Dish.csv')
df_dish

df_dish['price_var'] = df_dish.apply(lambda x: x['highest_price'] - x['lowest_price'] if x['lowest_price'] > 0 else 0, axis=1)

df_dish_sorted_price_var = df_dish.sort_values(by=['price_var'], ascending=False)
df_dish_price_var_sorted_top15 = df_dish_sorted_price_var.head(15).copy()

plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
sns.barplot(x=df_dish_price_var_sorted_top15.name, y=df_dish_price_var_sorted_top15.price_var);
```
