```python
#Menu Ds
df_menu = pd.read_csv('./Menu.csv')
df_menu
```

```python
columns = ['id','sponsor','physical_description','notes','date', 'location', 'status', 'page_count', 'dish_count']
df_menu_filtered = df_menu[columns]
df_menu_filtered_w_date = df_menu_filtered[df_menu_filtered.date.isna() == False]
```

```python
df_menu_filtered_w_date['date'] = pd.to_datetime(df_menu_filtered_w_date['date'],errors = 'coerce')
df_menu_filtered_w_date['Year'] = df_menu_filtered_w_date['date'].dt.year
df_menu_filtered_w_date
```

```python
#Menu Ds
df_dish = pd.read_csv('./Dish.csv')
df_dish
```

```python
df_dish['price_var'] = df_dish.apply(lambda x: x['highest_price'] - x['lowest_price'] if x['lowest_price'] > 0 else 0, axis=1)
df_dish['year_var'] = df_dish.apply(lambda x: x['last_appeared'] - x['first_appeared'], axis=1)
```

```python
#Menu Ds
df_menu_page = pd.read_csv('./MenuPage.csv')
df_menu_page
```

```python
#Menu Ds
df_menu_item = pd.read_csv('./MenuItem.csv')
df_menu_item
```

```python
#Merging DF
df_menuitem_dish = df_menu_item.merge(df_dish, how='left', left_on=['dish_id'], right_on=['id'])
df_menuitem_dish.rename(columns = {'id_x': 'id_menu_item','id_y': 'id_dish'}, inplace = True)

df_menuitem_menupage_dish = df_menuitem_dish.merge(df_menu_page, how='left', left_on=['menu_page_id'], right_on=['id'])
df_menuitem_dish.rename(columns = {'id': 'id_menu_page'}, inplace = True)

df_merged = df_menuitem_menupage_dish.merge(df_menu_filtered_w_date, how='left', left_on=['menu_id'], right_on=['id'])
```

```python
merged_columns = ['menu_page_id',
                  'id_menu_item',
                  'price',
                  'high_price',
                  'xpos',
                  'ypos',
                  'menus_appeared',
                  'times_appeared',
                  'first_appeared',
                  'last_appeared',
                  'lowest_price',
                  'highest_price',
                  'name',
                  'menu_id',
                  'dish_count',
                  'location',
                  'date',
                  'sponsor',
                  'image_id',
                  'full_height',
                  'full_width',
                  'physical_description',
                  'Year'
                 ]
df_merged_filtered = df_merged[merged_columns].copy()
```

```python
df_cp = df_merged_filtered[['sponsor', 'price', 'name', 'Year', 'last_appeared']]
df_cp['times_appeared_by_year'] = 1
groupped = df_cp.groupby(['Year', 'name']).sum().reset_index()
group_sort = groupped.sort_values(by=['times_appeared_by_year'], ascending=False)
```

```python
 df_ = group_sort[group_sort.name.isin(['Tea', 
                                       'Boiled potatoes' , 
                                       'Olives', 
                                       'Milk', 
                                       'Chocolate',
                                       'Coffee',
                                       'Celery',
                                       'Radishes',
                                       'Mashed potatoes',
                                       'Fruit',
                                       'Chicken salad',
                                       'Vanilla ice cream',
                                       'Cocoa',
                                       'Lettuce Salad',
                                       'Cheese',
                                       'Lobster salad'
                                      ])]
```
