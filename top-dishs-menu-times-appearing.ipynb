```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Menu Ds
df_dish = pd.read_csv('./Dish.csv')
df_dish
```

```python
df_dish_sorted = df_dish.sort_values(by=['times_appeared'], ascending=False)
df_dish_sorted_top15 = df_dish_sorted.head(15).copy()

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
#plt.title(schema.Country)
sns.barplot(x=df_dish_sorted_top15.name, y=df_dish_sorted_top15.times_appeared);
```

```python
df_dish_menu_sorted = df_dish.sort_values(by=['menus_appeared'], ascending=False)
df_dish_menu_sorted_top15 = df_dish_menu_sorted.head(15).copy()

plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x=df_dish_menu_sorted_top15.name, y=df_dish_menu_sorted_top15.menus_appeared);
```
